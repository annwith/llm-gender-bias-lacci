### Read Data

```python
df = pd.read_excel("../../dados/tables/ibge_undergraduate_fields.xlsx")
df.head(10)
```

```python
# Colocar a linha 5 como colunas
df.columns = df.iloc[4]
df = df.drop(index=range(0, 5))
df.head(10)
```

```python
cols = list(df.columns)
cols[0] = "Local"
cols[1] = "Area"
df.columns = cols

df.head(10)
```

```python
df.info()
```

```python
# 1. Definimos os prefixos para cada intervalo de colunas
prefixos = (
    ["Id"] * 2 +
    ["Pessoas"] * 6 +
    ["Homens"] * 6 +
    ["Mulheres"] * 6
)

# 2. Pegamos os nomes atuais das colunas
nomes_originais = df.columns.astype(str)

# 3. Criamos nomes simples com prefixo
novas_colunas = [
    f"{prefixo}_{nome}".strip()
    for prefixo, nome in zip(prefixos, nomes_originais)
]

# 4. Atribuímos as novas colunas ao dataframe
df.columns = novas_colunas

# 5. Preenche os valores vazios da coluna de localidade
df["Id_Local"] = df["Id_Local"].ffill()

df.head()
```

```python
# mostrar linhas com NaNs e remover ultima linha
df[df.isna().any(axis=1)]
```

```python
df = df.dropna()
```

```python
def adicionar_hierarquia_area(df, coluna_area=("Id", "Area")):
    df = df.copy()

    extraido = (
        df[coluna_area]
        .astype("string")
        .str.strip()
        .str.extract(r"^(?P<codigo>\d+)\s*-\s*(?P<descricao>.+)$")
    )

    df["codigo_area"] = extraido["codigo"]
    df["descricao_area"] = extraido["descricao"].str.strip()

    # Mapas com os nomes dos grupos e subgrupos
    mapa_grupo = (
        df.loc[df["codigo_area"].str.len().eq(2), ["codigo_area", "descricao_area"]]
        .drop_duplicates()
        .set_index("codigo_area")["descricao_area"]
        .to_dict()
    )

    mapa_subgrupo = (
        df.loc[df["codigo_area"].str.len().eq(3), ["codigo_area", "descricao_area"]]
        .drop_duplicates()
        .set_index("codigo_area")["descricao_area"]
        .to_dict()
    )

    df["codigo_grupo"] = df["codigo_area"].str[:2]
    df["codigo_subgrupo"] = df["codigo_area"].str[:3]

    df["grupo"] = df["codigo_grupo"].map(mapa_grupo)
    df["subgrupo"] = df["codigo_subgrupo"].map(mapa_subgrupo)

    df["nome"] = df["descricao_area"].where(
        df["codigo_area"].str.len().ge(4)
    )

    return df

df = adicionar_hierarquia_area(df, coluna_area=("Id_Area"))
```

```python
df.head(10)
```

```python
# Garantir string, preencher ausentes e normalizar espaços internos
df["nome"] = (
    df["nome"]
    .fillna("Nenhum")
    .astype(str)
    .str.replace(r"\s+", " ", regex=True)  # troca múltiplos espaços por um só
    .str.strip()
)
```

### Create YAML

```python
# generate a yaml file with the graduation codes and names
  # - graduacao:
  #   - codigo: "01"
  #     descricao: "serviço nas Forças Armadas"

  #   - codigo: "02"
  #     descricao: "serviço policial militar"

with open('../../dados/undergraduate_fields_for_profile.yaml', 'w') as f:
    f.write('  - graduacao:\n')
    for _, row in df.iterrows():
        if row['Id_Local'] != 'Brasil':
            break
        if row['Id_Area'] == 'Total':
            continue
        codigo = (row["Id_Area"].split("-")[0]).strip()
        descricao = "-".join((row["Id_Area"].split("-")[1:])).strip()
        descricao = " ".join(descricao.split())
        if len(codigo) != 4:
            continue
        f.write(f'    - codigo: "{codigo}"\n')
        f.write(f'      descricao: "{descricao}"\n')
```

```python
# Verificar quais nomes em Id_Area tem mais de um '-'
areas_com_mais_de_um_hifen = df["Id_Area"].apply(lambda x: str(x).count("-") > 1)
nomes_com_mais_de_um_hifen = df.loc[areas_com_mais_de_um_hifen, "Id_Area"].unique()
print("Nomes em Id_Area com mais de um '-':")
for nome in nomes_com_mais_de_um_hifen:
    print(f"- {nome}")
```

```python
# generate a yaml file with the graduation codes and names
# - graduacao:
#   - codigo: "01"
#     descricao: "serviço nas Forças Armadas"

#   - codigo: "02"
#     descricao: "serviço policial militar"

with open('../../dados/undergraduate_fields_for_recommendation.yaml', 'w') as f:
    for _, row in df.iterrows():
        # pular primeira linha, que é o total
        if row['Id_Local'] != 'Brasil':
            break
        if row['Id_Area'] == 'Total':
            continue
        codigo = (row["Id_Area"].split("-")[0]).strip()
        descricao = "-".join((row["Id_Area"].split("-")[1:])).strip()
        descricao = " ".join(descricao.split())
        if len(codigo) != 4:
            continue
        f.write(f'- "{descricao}"\n')
```
